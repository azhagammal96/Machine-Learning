```python
import os
import numpy as np
import matplotlib.pyplot as plt
import scipy.io
import pandas as pd
from sklearn.cluster import KMeans
```

```python
mat_dt = scipy.io.loadmat('merge_dt.mat')
mat_pia = scipy.io.loadmat('merge_pia.mat')
mat_stra = scipy.io.loadmat('merge_stra.mat')
```

```python
df_dt = pd.DataFrame(mat_dt['dfc'])
df_pia = pd.DataFrame(mat_pia['dfc'])
df_stra = pd.DataFrame(mat_stra['dfc'])
```

```python
#Finding optimal number of clusters

Error =[]
for i in range(1, 5):
    kmeans = KMeans(n_clusters = i).fit(df_pia)
    kmeans.fit(df_pia)
    Error.append(kmeans.inertia_)
import matplotlib.pyplot as plt
plt.plot(range(1, 5), Error)
plt.title('Elbow method for three pieces')
plt.xlabel('No of clusters')
plt.ylabel('Error')
plt.show()
```

```python
#Kmeans for Dream Theatre and subsequent plotting

kmeans_dt = KMeans(n_clusters = 2)
y_kmeans_dt = kmeans_dt.fit_predict(df_dt)
print(y_kmeans_dt)
clus_dt = kmeans_dt.cluster_centers_
print(clus_dt)

plt.scatter(df_dt.iloc[:, 0], df_dt.iloc[:, -1])
    
plt.scatter(kmeans_dt.cluster_centers_[:, 0], kmeans_dt.cluster_centers_[:, 1], c='red', marker='x')
    
plt.title('Data points and cluster centroids for Dream Theatre')
plt.show()
```

```python
#Kmeans for Piazzolo and subsequent plotting

kmeans_pia = KMeans(n_clusters = 2)
y_kmeans_pia = kmeans_pia.fit_predict(df_pia)
print(y_kmeans_pia)
clus_pia = kmeans_pia.cluster_centers_
print(clus_pia)

plt.scatter(df_pia.iloc[:, 0], df_pia.iloc[:, -1])
    
plt.scatter(kmeans_pia.cluster_centers_[:, 0], kmeans_pia.cluster_centers_[:, 1], c='red', marker='x')
    
plt.title('Data points and cluster centroids for Piazzolo')
plt.show()
```

```python
#Kmeans for Dream Theatre and subsequent plotting

kmeans_stra = KMeans(n_clusters = 2)
y_kmeans_stra = kmeans_stra.fit_predict(df_stra)
print(y_kmeans_stra)
clus_stra = kmeans_stra.cluster_centers_
print(clus_stra)

plt.scatter(df_stra.iloc[:, 0], df_stra.iloc[:, -1])
    
plt.scatter(kmeans_stra.cluster_centers_[:, 0], kmeans_stra.cluster_centers_[:, 1], c='red', marker='x')
    
plt.title('Data points and cluster centroids for Stravin')
plt.show()
```
